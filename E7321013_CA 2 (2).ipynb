```python
#2.
class Complex:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def __sub__(self, other):
        return self.a - other.a, self.b - other.b
        
class negative:
    def __init__(self, a):
        self.a = a
                
    def __neg__(self):
         return -self.a
    
print("""Choose the options:
                1. Difference between two complex numbers
                2. Negative of a number.""")


val = 0
while val == 0:
    option = int(input("Choose the option: "))
    if option == 1:
        obj1 = Complex(4, 5)
        obj2 = Complex(1, 2)
        e = obj1 - obj2
        print(e)
    
    elif option == 2:
        value = int(input("Enter the value: "))
        negs = negative(value)
        print(-negs)
    
    else:
        print("Something went wrong!")
    
    val = int(input("Enter 0 to continue the process or 1 to exit. "))
```

```python
#4
class Error(Exception):
    pass

class ZeroBalanceError(Error):
    pass

budgdict = [{"category": "Food", "amount": 0},
           {"category": "Clothing", "amount": 0},
           {"category": "Entertainment", "amount": 0}]

class Budget(object):
    
    def init(self):
        pass
    def depositer(self):
        print("Welcome!")
        print("""Deposit Amount for
                            1.Food
                            2.Clothing
                            3.Entertainment""")
        depositval = 0
        while depositval == 0:
            deposit_choice = int(input("Enter your choice: "))
            if deposit_choice >= 4:
                print("Enter a valid choice: ")
            else:
                print("Category Name: ", budgdict[deposit_choice - 1]["category"])
                print("Available balance: ", budgdict[deposit_choice - 1]["amount"])
                amountd = int(input("Enter the amount to deposit: "))
                budgdict[deposit_choice - 1]["amount"] += amountd

            depositval = int(input("Enter 0 to continue or enter 1: "))
    
    def withdrawer(self):
        print("Welcome Again!")
        print("""Withdraw Amount from
                            1.Food
                            2.Clothing
                            3.Entertainment""")
        withval = 0
        while withval == 0:
            withdraw_choice = int(input("Enter your choice: "))
            if withdraw_choice >= 4:
                print("Enter a valid choice: ")
            else:
                print("Category Name: ", budgdict[withdraw_choice - 1]["category"])
                print("Available balance: ", budgdict[withdraw_choice - 1]["amount"])
                amountw = int(input("Enter the amount to withdraw: "))
                try:
                    if amountw > budgdict[0]["amount"]:
                        raise ZeroBalanceError("You have less balance in your account.")
                except ZeroBalanceError:
                    print("You have less balance in your account.")
                else:
                    budgdict[withdraw_choice - 1]["amount"] = budgdict[withdraw_choice - 1]["amount"] - amountw

            withval = int(input("Enter 0 to continue or enter 1: "))
            
    def tranferer(self):
        print("Welcome Again!")
        print("""Transfer amount from
                            1.Food
                            2.Clothing
                            3.Entertainment""")
        category_from = int(input())
        print("""Transfer amount to
                            1.Food
                            2.Clothing
                            3.Entertainment""")
        category_to = int(input())
        print("Available balance in ",budgdict[category_from - 1]["amount"])
        print("Available balance in ",budgdict[category_to - 1]["amount"])
        tranfer_amount = int(input("Enter the amount to tranfer"))
        if category_from == category_to:
            print("You can't transfer the amount to the same category")
        else:
            budgdict[category_from - 1]["amount"] = budgdict[category_from - 1]["amount"] - tranfer_amount 
            budgdict[category_to - 1]["amount"] += tranfer_amount 
            
a = Budget()
a.depositer()
a.withdrawer()
a.tranferer()
print(budgdict)
```
