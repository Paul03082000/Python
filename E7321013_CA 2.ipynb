```python
class Error(Exception):
    pass

class NOgradeError(Error):
    pass

class Subject_list(object):
    
    class_list = []
    
    def __init__(self, test_grades, test_marks):
        self.test_grades = test_grades
        self.test_marks = test_marks
        
    def find_average(self, test_marks):
        try:
            if test_marks == []:
                raise NOgradeError("NO GRADE FOR A STUDENT")
        except NOgradeError:
            print("NO GRADE FOR A STUDENT.")
        else:
            avege = sum(test_marks) / len(test_marks)
            return avege
    
    def displayer(self):
#         Subject_list.find_average(self, test_marks)
        Subject_list.class_list.append(self.test_grades)
        print("Class List = ", Subject_list.class_list)  
        
ini = 0
names = []
marks = []
test_grades = []
while ini == 0:
    dummy = []
    names = list(map(str, input("Enter the name: ").split()))
    marks = list(map(float, input("Enter the marks: ").split()))
    test_names = [n for n in names] 
    test_marks = [m for m in marks]
    dummy.append(test_names)
    dummy.append(test_marks)
    averager = Subject_list.find_average(test_grades, test_marks)
    dummy.append(averager)
    test_grades.append(dummy)
    mark_list = Subject_list(test_grades, test_marks)
    ent = int(input("Enter 0 if you want to add more marklists if not enter 1: "))
    if ent != 0:
        ini += 1

mark_list.displayer()
```

```python
class Complex:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def __sub__(self, other):
        return self.a - other.a, self.b - other.b
        
class negative:
    def __init__(self, a):
        self.a = a
                
    def __neg__(self):
         return -self.a
    
print("""Choose the options:
                1. Difference between two complex numbers
                2. Negative of a number.""")


val = 0
while val == 0:
    option = int(input("Choose the option: "))
    if option == 1:
        obj1 = Complex(4, 5)
        obj2 = Complex(1, 2)
        e = obj1 - obj2
        print(e)
    
    elif option == 2:
        value = int(input("Enter the value: "))
        negs = negative(value)
        print(-negs)
    
    else:
        print("Something went wrong!")
    
    val = int(input("Enter 0 to continue the process or 1 to exit. "))
```

```python
class Error(Exception):
    pass

class ZeroBalanceError(Error):
    pass

class Budget(object):
    def __init__(self):
        deposit_food = 0
        deposit_clothing = 0
        deposit_entertainment = 0
        withdrawn_food = 0
        withdrawn_clothing = 0
        withdrawn_entertainment = 0
        
    def zero_bal(self, dep, wit):
        try:
            if wit > dep:
                raise ZeroBalanceError("You have less balance in your account.")
        except ZeroBalanceError:
            print("You have less balance in your account.")
        else:
            return dep - wit
    
    class Food(Budget):
        def __init__(self, deposit_food, withdrawn_food):
            super(Food, self).__init__()
            self.deposit_food = deposit_food
            self.withdrawn_food = withdrawn_food
    
    class Clothing(Budget):
        def __init__(self, deposit_clothing, withdrawn_clothing):
            super(Clothing, self).__init__()
            self.deposit_clothing = deposit_clothing
            self.withdrawn_clothing = withdrawn_clothing
            
    class Entertainment(Budget):
        def __init__(self, deposit_entertainment, withdrawn_entertainment):
            super(Entertainment, self).__init__()
            self.deposit_entertainment = deposit_entertainment
            self.withdrawn_entertainment = withdrawn_entertainment
                
    def budget_displayer(self):
        print("Deposited Amount for ")
        print("Food: ", deposit_food)
        print("Clothing: ", deposit_clothing)
        print("Entertainment: ", deposit_entertainment)
        print("Withdrawing amount: ")
        print("Food: ", withdrawn_food)
        print("Clothing: ", withdrawn_clothing)
        print("Entertainment: ", withdrawn_entertainment)
    
print("BUDGET CALCULATOR.")

print("""Calculate budget for:
                        1.Food
                        2.Clothing
                        3.Entertainment""")

value = 0
while value == 0:
    option = int(input("Enter your option: "))
    if option == 1:
        deposit_food = int(input("Enter the deposit amount: "))
        withdrawn_food = int(input("Enter the withdraw amount: "))
        food = Food(deposit_food, withdrawn_food)
    
    elif option == 2:
        deposit_clothing = int(input("Enter the deposit amount: "))
        withdrawn_clothing = int(input("Enter the withdraw amount: "))
        clothing = Clothing(deposit_clothing, withdrawn_clothing)
    
    elif option == 3:
        deposit_entertainment = int(input("Enter the deposit amount: "))
        withdrawn_entertainment = int(input("Enter the withdraw amount: "))
        entertainment = Entertainment(deposit_entertainment, withdrawn_entertainment)
        
    else:
        print("Enter a valid option.")
        
    value = int(input("Enter 0 to Continue or enter 1: "))
    
final = Budget()
final.budget_displayer()
foodb = final.zero_bal(deposit_food, withdrawn_food)
clothesb = final.zero_bal(deposit_clothing, withdrawn_clothing)
entertainmentb = final.zero_bal(deposit_entertainment, withdrawn_entertainment)

print("BALANCE AVAILABLE:")
print("Balance in food category: ", foodb)
print("Balance in clothing category: ", clothesb)
print("Balance in entertainment: ", entertainmentb)
```
